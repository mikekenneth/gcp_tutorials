```python
import os
from google.cloud import pubsub_v1
```

```python
# To get the credentials.json file https://cloud.google.com/iam/docs/creating-managing-service-account-keys#creating
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/home/kenneth/tutorials/google_cloud_tutorials/credentials.json"

# TODO (developer): Set the PROJECT_ID & TOPIC
project_id = "qwiklabs-gcp-01-bf5ecf8a32ef"
topic = "sandiego"
```

```python
topic_name = f'projects/{project_id}/topics/{topic}'
subscription_name = f'projects/{project_id}/subscriptions/mysub_2'

# Define the Callback (action to be taken for each messages)
def callback(message):
    print(message.data)
    message.ack()

with pubsub_v1.SubscriberClient() as subscriber:
    # subscriber.create_subscription(name=subscription_name, topic=topic_name)
    future = subscriber.subscribe(subscription_name, callback)
    try:
        future.result()
    except KeyboardInterrupt:
        future.cancel()
```
